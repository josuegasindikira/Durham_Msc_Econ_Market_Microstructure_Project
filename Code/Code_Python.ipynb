```python
#GRAPHIQUES VARAIBLES ACROSS REGIMES
import pandas as pd

import os
os.getcwd()

df = pd.read_csv(r"C:\Users\Dell\Desktop\Courses\Market Microstructure\Summarative\data1graph.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import StrMethodFormatter
import math

# =========================================================
# 1. USER-DEFINED INPUTS
# =========================================================

date_col = "auction_date"

# Optional alternative set of variables
#variables_to_plot = [
 #   "bid_amount",
   # "amount_served",
    #"bid_to_cover",
    #"outstand_notional",
    #"Vix_euro"
#]

# Main set of variables to display
variables_to_plot = [
    "bid_amount",
    "amount_served",
    #"bid_to_cover",
    "outstanding_notional",
    #"Vix_euro",
    "auction_yield",
    "weighted_average_price",
    "benchmark_yield",
    "auction_premium_bp",
    "shock_log_dev",
]

# Additional variables reserved for a separate figure
variables_to_plot2 = [
   "bid_to_cover",
    "Vix_euro",
]


n_rows = 2
n_cols = 4

subplot_width = 7
subplot_height = 4

date_format = "%m/%d/%Y"

# =========================================================
# 2. DATA PREPARATION
# =========================================================

# Convert the auction date column to datetime format
df[date_col] = pd.to_datetime(df[date_col], format=date_format, errors="coerce")

# Convert selected plotting variables to numeric format
for var in variables_to_plot:
    df[var] = pd.to_numeric(df[var], errors="coerce")

# Remove observations with missing auction dates
df = df.dropna(subset=[date_col]).copy()

# Sort observations chronologically
df = df.sort_values(date_col)

# =========================================================
# 3. DEFINE CRISIS PERIODS
# =========================================================

# Global Financial Crisis period
gfc_start = pd.Timestamp("2008-07-01")
gfc_end   = pd.Timestamp("2009-04-30")

# COVID-19 crisis period
covid_start = pd.Timestamp("2020-02-02")
covid_end   = pd.Timestamp("2021-12-31")

# Create equally long pre-crisis and post-crisis comparison windows
def same_length_windows(start, end):
    n_days = (end - start).days + 1

    before_end = start - pd.Timedelta(days=1)
    before_start = before_end - pd.Timedelta(days=n_days - 1)

    after_start = end + pd.Timedelta(days=1)
    after_end = after_start + pd.Timedelta(days=n_days - 1)

    return before_start, before_end, after_start, after_end

# Generate comparison windows around the GFC period
gfc_before_start, gfc_before_end, gfc_after_start, gfc_after_end = same_length_windows(gfc_start, gfc_end)

# Generate comparison windows around the COVID-19 period
covid_before_start, covid_before_end, covid_after_start, covid_after_end = same_length_windows(covid_start, covid_end)

# =========================================================
# 4. CREATE PERIOD INDICATORS
# =========================================================

# GFC period indicators
df["bool_gfc_before"] = df[date_col].between(gfc_before_start, gfc_before_end)
df["bool_gfc_during"] = df[date_col].between(gfc_start, gfc_end)
df["bool_gfc_after"]  = df[date_col].between(gfc_after_start, gfc_after_end)

# COVID-19 period indicators
df["bool_covid_before"] = df[date_col].between(covid_before_start, covid_before_end)
df["bool_covid_during"] = df[date_col].between(covid_start, covid_end)
df["bool_covid_after"]  = df[date_col].between(covid_after_start, covid_after_end)

# =========================================================
# 5. HELPER FUNCTION TO COMPUTE PERIOD MEANS
# =========================================================

# Compute the mean of a variable over a selected period
def get_mean(data, var, bool_col):
    return data.loc[data[bool_col], var].mean()

# =========================================================
# 6. CREATE THE MULTI-PANEL FIGURE
# =========================================================

# Initialize the figure and subplot grid
fig, axes = plt.subplots(
    n_rows,
    n_cols,
    figsize=(n_cols * subplot_width, n_rows * subplot_height)
)

# Convert the axes object into an iterable structure
if n_rows == 1 and n_cols == 1:
    axes = [axes]
else:
    axes = axes.flatten()

# =========================================================
# 7. LOOP OVER VARIABLES AND CREATE EACH SUBPLOT
# =========================================================

for i, var in enumerate(variables_to_plot):

    # Stop if the number of variables exceeds the available panels
    if i >= len(axes):
        break

    ax = axes[i]

    # Keep only non-missing observations for the current variable
    temp = df.dropna(subset=[var]).copy()

    # Plot the main time-series line
    ax.plot(temp[date_col], temp[var], color="steelblue", linewidth=0.9)

    # Highlight crisis periods with shaded areas
    ax.axvspan(gfc_start, gfc_end, color="red", alpha=0.12)
    ax.axvspan(covid_start, covid_end, color="orange", alpha=0.12)

    # Compute means before, during, and after the GFC
    mean_gfc_before = get_mean(temp, var, "bool_gfc_before")
    mean_gfc_during = get_mean(temp, var, "bool_gfc_during")
    mean_gfc_after  = get_mean(temp, var, "bool_gfc_after")

    # Compute means before, during, and after COVID-19
    mean_covid_before = get_mean(temp, var, "bool_covid_before")
    mean_covid_during = get_mean(temp, var, "bool_covid_during")
    mean_covid_after  = get_mean(temp, var, "bool_covid_after")

    # Compute a vertical offset for mean labels
    y_valid = temp[var].dropna()
    if len(y_valid) > 0:
        y_span = y_valid.max() - y_valid.min()
        offset = 0.03 * y_span if y_span > 0 else 1
    else:
        offset = 1

    # Add a horizontal mean segment and its corresponding label
    def add_mean_segment(ax, start, end, mean_value, label):
        if pd.notna(mean_value):
            ax.plot(
                [start, end],
                [mean_value, mean_value],
                color="black",
                linewidth=1.0,
                solid_capstyle="butt"
            )

            mid = start + (end - start) / 2
            ax.text(
                mid,
                mean_value + offset,
                label,
                ha="center",
                va="bottom",
                fontsize=10,
                color="black"
            )

    # Add GFC mean segments
    add_mean_segment(ax, gfc_before_start, gfc_before_end, mean_gfc_before, "before")
    add_mean_segment(ax, gfc_start, gfc_end, mean_gfc_during, "gfc")
    add_mean_segment(ax, gfc_after_start, gfc_after_end, mean_gfc_after, "after")

    # Add COVID-19 mean segments
    add_mean_segment(ax, covid_before_start, covid_before_end, mean_covid_before, "before")
    add_mean_segment(ax, covid_start, covid_end, mean_covid_during, "covid")
    add_mean_segment(ax, covid_after_start, covid_after_end, mean_covid_after, "after")

    # Set subplot title and axis labels
    ax.set_title(var, fontsize=12)
    ax.set_xlabel("time", fontsize=10)
    ax.set_ylabel(var, fontsize=10)

    # Format the x-axis as calendar years
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    ax.tick_params(axis="x", rotation=45)

    # Format y-axis values and add a reference grid
    ax.yaxis.set_major_formatter(StrMethodFormatter("{x:,.0f}"))
    ax.grid(True, linestyle="--", alpha=0.4)

# =========================================================
# 8. HIDE UNUSED SUBPLOTS
# =========================================================

# Hide empty panels when fewer variables than subplots are used
for j in range(len(variables_to_plot), len(axes)):
    axes[j].set_visible(False)

# Adjust spacing, save the figure, and display it
plt.tight_layout()
plt.savefig("mon_graph4.png", dpi=300, bbox_inches="tight")
plt.show()
```
